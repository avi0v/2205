```python
import pandas as pd 
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
#loading data
titanic=pd.read_csv("Titanic_Apriori.csv")
titanic.head()
```

```python
df=pd.get_dummies(titanic,dtype=int)
df
```

```python
df.info()
```

```python
#building model

freq_itemset=apriori(df,min_support=0.5,use_colnames=True)
freq_itemset
```

```python
#
rules=association_rules(freq_itemset,metric="lift",min_threshold=0.7)
```

```python
rules
```

```python
comb=rules.loc[:, rules.columns.isin(['antecedents','consequents','lift'])]
```
